```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('train.csv', index_col='ID')
train_data.target.unique()
```

```python
train_data.shape
```

```python
n = train_data.shape[0]
n1 = np.sum(train_data["target"])
print("Number of lines: {}".format(n))
print("Percentage of 1: {}".format(n1/n))
print("Percentage of 0: {}". format(1-n1/n))
```

```python
columns_full = []

for c in train_data.columns:
    p_nan = train_data[c].isnull().mean()
    print("Column {0}:\n {1}NaN variables".format(c, p_nan))
    if p_nan < 0.2:
        columns_full.append(c)

len(columns_full)  # dont la target
```

```python
nonnum_cols = {}
for i in range(1, 132):
    if type(train_data['v' + str(i)].iloc[0]) != np.float64:
        nonnum_cols['v' + str(i)] = train_data['v' + str(i)].unique()
```

```python
cont_cols = list(set(train_data.columns) - set(nonnum_cols.keys()))
cont_data = train_data[cont_cols]

for c in cont_cols:
    sns.distplot(cont_data[cont_data.target == 0][c].dropna(), color='red', bins=15)
    sns.distplot(cont_data[cont_data.target == 1][c].dropna(), color='blue', bins=15)
    plt.show()
```
